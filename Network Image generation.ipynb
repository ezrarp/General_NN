```python
import matplotlib.pyplot as plt
class Neuralimage :
    def __init__(self, structure, weights, bias) :
        self.structure = structure
        self.weights = weights
        self.bias = bias
        self.image = self.generate_image(structure, weights, bias)

    def generate_image(self, structure: list, weights:list, bias: int):
        layer_x_val = [i + 1 for i in range(len(structure))]
        layer_y_val = []
        for i in range(len(structure)):
            layer_y_val.append([j + 1 for j in range(structure[i])])

        for x in layer_x_val:
            for y in layer_y_val[x - 1]:
                plt.scatter(x, y, color='lightblue', s = 800, alpha=1)
                plt.text(x, y, f'{y}', fontsize=12, ha='right')

        line = 0 
        for x in range(len(layer_x_val)):
            for y in range(len( layer_y_val[x])):
                if x + 1 < len(layer_x_val):
                    for z in range(len(layer_y_val[x + 1])):

                        plt.plot([x + 1, x + 2], [layer_y_val[x][y], layer_y_val[x + 1][z]], color='black', alpha=0.5, linewidth=2)
                        plt.text((x + 1 + x + 2) / 2, (layer_y_val[x][y] + layer_y_val[x + 1][z]) / 2,  'W' + str(line)+ '=' + f'{weights[line]}', fontsize=12, ha='right')
                        line += 1
                    
            

        plt.title('Neural Network')
        plt.xlabel('Layer')
        plt.ylabel('Neuron')
        plt.show()
        

        
        
```

```python
#test image generation
structure = [3, 9, 2,1]
weights = [ rand for rand in range(1, 61)]
bias = [1, 1]

image = Neuralimage(structure, weights, bias)
print(image.image)
```
